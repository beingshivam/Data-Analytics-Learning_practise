#  Image Manipulation using Numpy

```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img_array = ndimage.imread('datasets/koala.jpg')
```

```python
img_array.shape
# 3 D array
#3 represents R G B values of each pixel
```

```python
type(img_array)
```

```python
img_array
```

```python
#viewing image using matplotlib

plt.imshow(img_array)
plt.show()
```

```python
#image slicing

image_slice = img_array[300:,250:,:]
```

```python
image_slice
```

```python
#viewing image

plt.imshow(image_slice)
plt.show()
```

```python
# image splitting along x-axis

x1 , x2 = np.split(img_array, 2)
```

```python
plt.imshow(x1)
plt.show()
```

```python
plt.imshow(x2)
plt.show()
```

```python
# image splitting along y axis

y1 , y2 = np.split(img_array, 2, axis=1)
```

```python
plt.imshow(y1)
plt.show()
```

```python
plt.imshow(y2)
plt.show()
```

```python
#combining images using numpy's concatenate function

plt.imshow(np.concatenate((x1,x2)))
plt.show()
```

```python
# combining along y axis

plt.imshow(np.concatenate((y1,y2)))
plt.show()
```

```python
#specify along y axis

plt.imshow(np.concatenate((y1,y2),axis=1))
plt.show()
```
